```python
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------------------------------------
# Runtime Scaling Plot (Log-Log)
# ----------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt

# --- Data ---
n = np.array([10, 50, 100, 250, 500, 1000, 2000, 3000, 4000, 5000])
time = np.array([0.075, 0.36, 0.74, 2.00, 4.50, 9.80, 18.50, 30.40, 56.70, 82.30])



# --- Fit log-log line ---
log_n = np.log10(n)
log_t = np.log10(time)
coeffs = np.polyfit(log_n, log_t, 1)
slope, intercept = coeffs

# --- Plot ---
plt.figure(figsize=(6, 4))
plt.scatter(log_n, log_t, color='blue', label='Data points', zorder=3)
plt.plot(log_n, np.polyval(coeffs, log_n), color='red',
         label=f'Fit: log(T) = {slope:.2f}·log(n) + {intercept:.2f}', zorder=2)

# --- Style ---
plt.xlabel(r'$\log_{10}(n)$', fontsize=12)
plt.ylabel(r'$\log_{10}(\mathrm{Time~[s]})$', fontsize=12)
plt.title('Log–Log Plot of Runtime Scaling', fontsize=13)
plt.grid(True, which='both', linestyle='--', linewidth=0.6)
plt.legend(fontsize=10)
plt.tight_layout()

# --- Show slope in console ---
print(f"Estimated complexity exponent p = {slope:.2f}")
plt.show()
```

```python
# ----------------------------------------------------------
# Publication-Ready Log–Log Plot of Runtime Scaling
# ----------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt

# --- Data ---
n = np.array([10, 50, 100, 250, 500, 1000, 2000, 3000, 4000, 5000])
time = np.array([0.075, 0.36, 0.74, 2.00, 4.50, 9.80, 18.50, 30.40, 56.70, 82.30])

# --- Fit log-log line ---
log_n = np.log10(n)
log_t = np.log10(time)
slope, intercept = np.polyfit(log_n, log_t, 1)

# --- Global style settings (for journal consistency) ---
plt.rcParams.update({
    "font.family": "Times New Roman",
    "font.size": 11,
    "axes.linewidth": 0.8,
    "axes.labelsize": 12,
    "xtick.direction": "in",
    "ytick.direction": "in",
    "xtick.major.size": 4,
    "ytick.major.size": 4,
    "xtick.minor.size": 2,
    "ytick.minor.size": 2,
    "legend.frameon": False,
})

# --- Plot ---
fig, ax = plt.subplots(figsize=(3.2, 2.4))  # small but crisp

# Scatter and regression line
ax.scatter(log_n, log_t, color='black', s=25, label='Data points', zorder=3)
ax.plot(log_n, np.polyval([slope, intercept], log_n),
        color='darkred', linewidth=1.2,
        label=fr'Fit: $\log(T) = {slope:.2f}\log(n) + {intercept:.2f}$', zorder=2)

# --- Axes labels and title ---
ax.set_xlabel(r'$\log_{10}(n)$')
ax.set_ylabel(r'$\log_{10}(\mathrm{Time~[s]})$')
# ax.set_title('Runtime Scaling', fontsize=12, pad=4)  # optional

# --- Grid and ticks ---
ax.grid(True, which='major', linestyle='--', linewidth=0.4, alpha=0.7)
ax.minorticks_on()

# --- Legend ---
ax.legend(loc='upper left', fontsize=6, handlelength=2)

# --- Layout ---
plt.tight_layout(pad=0.5)
plt.savefig("RuntimeScaling_LogLog.png", dpi=600, bbox_inches='tight')
plt.show()

print(f"Estimated complexity exponent p = {slope:.2f}")


```
